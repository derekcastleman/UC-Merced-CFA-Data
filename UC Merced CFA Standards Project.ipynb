## Comparing Standards

To compare retested standards, the different CFAs will be merged together with a full outer merge so that standards are not lost amongst tests. Standards will be compared going down the tests by seeing if there is a value in each column between the recent tests and working downward to previous tests. If there is a value common amongst test then the difference will be taken to see any change in percent correct for the standards.

At the end, the data generated in this section will be merged with the data frame from the previous section so that there is a new column that is added for Growth in standards between tests.

```python
if cfas == 2:
    cfa_one = pd.concat([ela_one, math_one]) #Concating tests together
    cfa_one.rename(columns = {'Test':'CFA One', 'Level': 'CFA One Level'}, inplace = True) #Renaming columns
    cfa_two = pd.concat([ela_two, math_two])
    cfa_two.rename(columns = {'Test':'CFA Two', 'Level': 'CFA Two Level'}, inplace = True)
if cfas == 3:
    cfa_one = pd.concat([ela_one, math_one])
    cfa_one.rename(columns = {'Test':'CFA One', 'Level': 'CFA One Level'}, inplace = True)
    cfa_two = pd.concat([ela_two, math_two])
    cfa_two.rename(columns = {'Test':'CFA Two', 'Level': 'CFA Two Level'}, inplace = True)
    cfa_three = pd.concat([ela_three, math_three])
    cfa_three.rename(columns = {'Test':'CFA Three', 'Level': 'CFA Three Level'}, inplace = True)
if cfas == 4:
    cfa_one = pd.concat([ela_one, math_one])
    cfa_one.rename(columns = {'Test':'CFA One', 'Level': 'CFA One Level'}, inplace = True)
    cfa_two = pd.concat([ela_two, math_two])
    cfa_two.rename(columns = {'Test':'CFA Two', 'Level': 'CFA Two Level'}, inplace = True)
    cfa_three = pd.concat([ela_three, math_three])
    cfa_three.rename(columns = {'Test':'CFA Three', 'Level': 'CFA Three Level'}, inplace = True)
    cfa_four = pd.concat([ela_four, math_four])
    cfa_four.rename(columns = {'Test':'CFA Four', 'Level': 'CFA Four Level'}, inplace = True)
else:
    pass
```

```python
# Creates a function thagt first looks if there is a value and then assigns one if present
if cfas == 2:
    def f(row):
        if row['CFA Two Level'] - row['CFA One Level'] >= -100: #Checks for a value being generated
            val = row['CFA Two Level'] - row['CFA One Level'] # Assigns the value
        else:
            val = '' # Gives no value if there is no crossover
        return val
    cfa_two = merge.loc[merge['CFA Two'] == 'CFA Two'] # Selects for only CFA Two
    cfa_two['Growth'] = cfa_two.apply(f, axis=1) # Applies the function a growth category
if cfas == 3:
    def f(row):
        if row['CFA Two Level'] - row['CFA One Level'] >= -100:
            val = row['CFA Two Level'] - row['CFA One Level']
        else:
            val = ''
        return val
    cfa_two = merge.loc[merge['CFA Two'] == 'CFA Two']
    cfa_two['Growth'] = cfa_two.apply(f, axis=1)
    def f(row):
        if row['CFA Three Level'] - row['CFA Two Level'] >= -100:
            val = row['CFA Three Level'] - row['CFA Two Level']
        elif row['CFA Three Level'] - row['CFA One Level'] >= -100: # Checks if it was on CFA 1 if it wasn't on 2
            val = row['CFA Three Level'] - row['CFA One Level']
        else:
            val = ''
        return val
    cfa_three = merge.loc[merge['CFA Three'] == 'CFA Three']
    cfa_three['Growth'] = cfa_three.apply(f, axis=1)
if cfas == 4:
    def f(row):
        if row['CFA Two Level'] - row['CFA One Level'] >= -100:
            val = row['CFA Two Level'] - row['CFA One Level']
        else:
            val = ''
        return val
    cfa_two = merge.loc[merge['CFA Two'] == 'CFA Two']
    cfa_two['Growth'] = cfa_two.apply(f, axis=1)
    def f(row):
        if row['CFA Three Level'] - row['CFA Two Level'] >= -100:
            val = row['CFA Three Level'] - row['CFA Two Level']
        elif row['CFA Three Level'] - row['CFA One Level'] >= -100:
            val = row['CFA Three Level'] - row['CFA One Level']
        else:
            val = ''
        return val
    cfa_three = merge.loc[merge['CFA Three'] == 'CFA Three']
    cfa_three['Growth'] = cfa_three.apply(f, axis=1)
    def f(row):
        if row['CFA Four Level'] - row['CFA Three Level'] >= -100:
            val = row['CFA Four Level'] - row['CFA Three Level']
        elif row['CFA Four Level'] - row['CFA Two Level'] >= -100:
            val = row['CFA Four Level'] - row['CFA Two Level']
        elif row['CFA Four Level'] - row['CFA One Level'] >= -100:
            val = row['CFA Four Level'] - row['CFA One Level']
        else:
            val = ''
        return val
    cfa_four = merge.loc[merge['CFA Four'] == 'CFA CFA Four']
    cfa_four['Growth'] = cfa_four.apply(f, axis=1)
else:
    pass
```
